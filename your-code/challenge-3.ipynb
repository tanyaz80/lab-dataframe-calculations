```python
groep_orders=orders.groupby(['CustomerID']).sum().reset_index()
groep_orders #4339 customers
```

```python
range_95 = np.quantile(groep_orders['amount_spent'], 0.95, axis=0)
range_95  #5840.18                                         
```

```python
range_75 = np.quantile(groep_orders['amount_spent'], 0.75, axis=0)
range_75 #1661.64
```

```python
VIP=groep_orders.loc[(groep_orders['amount_spent']>range_95)]
VIP #(217 rows)
```

```python
list_VIP=list(np.unique(VIP['CustomerID'].values))

list_VIP
```

```python
Preffered=groep_orders.loc[(groep_orders['amount_spent']>range_75)& (groep_orders['amount_spent']<range_95)]
Preffered #(868 rows)
```

```python
list_Pr=list(np.unique(Preffered['CustomerID'].values))

list_Pr
```
